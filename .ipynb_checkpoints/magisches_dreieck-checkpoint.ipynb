```python
from sympy import *
```

```python
p= 5
s = 2
n = p**s + 1
```

```python
print n
```

```python
values = {}
values[1] = [symbols('x_{}'.format(i)) for i in range(1, n + 1)]
```

```python
values
```

```python
for i in range(2, n + 1):
    values[i] = [p - values[i - 1][j] + p - values[i - 1][j + 1] for j in range(0, n - i + 1)]
```

```python
values
```

```python
res = 0
for s in values[1]:
    res += (values[n][0].coeff(s) % p) * s
```
