```python
from scipy.special import gamma
import math
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline    
```

```python
def fun(s):
    return 1.0 - np.sin(np.pi * gamma(s) / s) / np.sin(np.pi / s)
```

```python
xs = np.arange(10.9, 11.1, 0.001)   # start,stop,step
ys = [fun(x) for x in xs]
```

```python
plt.plot(xs,ys)
plt.show()
```
